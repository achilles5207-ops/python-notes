### Question 1: Basic Input/Output
Write a program that takes a user's name and age, and prints a greeting message. Calculate the year they will turn 100.

```
name = 'TestUser' # input('Enter name: ')
age = 20 # int(input('Enter age: '))
current_year = 2024
year_100 = current_year + (100 - age)
print(f'Hello {name}, you will turn 100 in the year {year_100}.')
```

### Question 2: Conditionals
Write a program to check if a given number is positive, negative, or zero.

```
num = 15
if num > 0:
    print('Positive')
elif num < 0:
    print('Negative')
else:
    print('Zero')
```

### Question 3: Looping
Write a Python program to print the Fibonacci sequence up to $N$ terms.

```
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=' ')
    a, b = b, a + b
print()
```

## 🚀 Pro-Tip: The `else` clause in Loops
In Python, you can use `else` with a `for` or `while` loop. The `else` block executes ONLY if the loop finishes without hitting a `break` statement.

### Question 4: Prime Number Check
Use a `for-else` loop to check if a number is prime.

```
num = 29
for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        print(f'{num} is not prime.')
        break
else:
    print(f'{num} is a prime number!')
```
