## 💡 Tip: The `with` statement
Always use the `with` context manager when opening files. It ensures the file is closed automatically, even if an exception occurs.

### Question 1: Reading and Writing Files
Write a list of strings to a file, then read the file line by line and print them.

```
lines = ['Python is fun\n', 'Data Science is cool\n']
with open('sample.txt', 'w') as f:
    f.writelines(lines)

with open('sample.txt', 'r') as f:
    for line in f:
        print(line.strip())

import os
os.remove('sample.txt') # clean up
```

### Question 2: Pickling (Binary Serialization)
Pickle a dictionary to a binary file, and then load it back.

```
import pickle
import os
data = {'user': 'Alice', 'score': 99}

# Save
with open('data.pkl', 'wb') as f:
    pickle.dump(data, f)

# Load
with open('data.pkl', 'rb') as f:
    loaded = pickle.load(f)
print(loaded)
os.remove('data.pkl') # clean up
```

## 🚀 Pro-Tip: Commit Transactions
In database programming, DML queries (INSERT, UPDATE, DELETE) are not permanently saved until you call `connection.commit()`.

### Question 3: SQLite Database Operations
Create an in-memory SQLite database, create a table `Books`, insert one book, and fetch it.

```
import sqlite3

conn = sqlite3.connect(':memory:') # Using in-memory DB for practice
cursor = conn.cursor()

cursor.execute('''CREATE TABLE Books (id INTEGER PRIMARY KEY, title TEXT, author TEXT)''')
cursor.execute("INSERT INTO Books (title, author) VALUES ('1984', 'George Orwell')")
conn.commit()

cursor.execute("SELECT * FROM Books")
print(cursor.fetchall())
conn.close()
```
