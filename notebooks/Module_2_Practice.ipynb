## 💡 Tip: List Comprehensions
List comprehensions are faster than regular `for` loops for creating lists because they are optimized in C underlying Python.

### Question 1: List Comprehension
Create a list of squares for all even numbers from 1 to 20.

```
even_squares = [x**2 for x in range(1, 21) if x % 2 == 0]
print(even_squares)
```

### Question 2: Dictionary Manipulation
Given a string, count the frequency of each character using a dictionary.

```
text = 'hello python'
freq = {}
for char in text:
    freq[char] = freq.get(char, 0) + 1
print(freq)
```

## 🚀 Pro-Tip: The Default Mutable Argument Trap
Never use a mutable object (like `[]` or `{}`) as a default argument in a function. Use `None` instead. Otherwise, the object is shared across all function calls.

### Question 3: Functions and `*args`
Write a function that accepts any number of numeric arguments and returns their product.

```
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(f'Product of 2,3,4,5 is: {multiply_all(2, 3, 4, 5)}')
```

### Question 4: Lambda Functions
Sort a list of tuples `[(name, age), ...]` based on age in descending order using a lambda function.

```
people = [('Alice', 25), ('Bob', 30), ('Charlie', 22)]
people.sort(key=lambda x: x[1], reverse=True)
print(people)
```
